### Histórico de venda de jogos de videogame 

```python
#Importando bibliotecas que serão usadas
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
#Lendo o arquivo
games = pd.read_csv('vgsales.csv')
```

```python
#Printando as primeiras 10 linhas da tabela
games.head(10)
```

```python
#tipos de dados em cada coluna
games.dtypes
```

```python
#Quantidade de linhas e colunas respectivamente
games.shape
```

```python
#Renomear colunas
games.columns = ['Ranking', 'Nome', 'Plataforma', 'Ano',
'Gênero', 'Editora', 'Vendas América do Norte', 'Vendas EUA',
'Vendas Japão', 'Outras vendas', 'Vendas Global']
```

```python
#Mostrar as 10 primeiras linhas do arquivo
games.head(10)
```

```python
#Verificando se há valores nulos
games.isnull().sum()
```

```python
#Contar jogos lançados por plataforma
games['Plataforma'].value_counts()
```

```python
#Mostrar resultado acima graficamente

lancados = games['Plataforma'].value_counts().head(10).plot(kind='bar', figsize=(11,5), grid = True, rot=0, color='blue')
plt.title('Os 10 videogames com mais títulos lançados')
plt.xlabel('Videogame') #Nomeando eixo X
plt.ylabel('Quantidade de jogos lançados') #Nomeando eixo Y
plt.show() 
```
